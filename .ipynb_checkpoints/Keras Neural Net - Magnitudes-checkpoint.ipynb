```python
import csv
filename = "TrSkew.csv"
fields = []
one = []
two = []
three = []
four = []
five = []
Y = []
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
     
    # extracting field names through first row
    fields =csvreader.next()
    for row in csvreader:
        one.append(float(row[3]))
        two.append(float(row[4]))  
        three.append(float(row[5]))
        four.append(float(row[6]))
        five.append(float(row[7])) 
        Y.append(float(row[9]))
```

```python
import numpy as np
X = np.column_stack((one,two,three,four,five))
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
np.random.seed(7)
```

```python
X
```

```python
model = Sequential()
model.add(Dense(6, input_dim=5, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X)
# round predictions
t = [round(x[0]) for x in predictions]
trainpred = np.array(t)
print trainpred
```

```python
correct = np.array(Y)
print correct
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(correct, trainpred)
```
